```python
import geopandas as gpd
gdf = gpd.read_file("mapData_20250129/TOWN_MOI_1131028.shp", encoding="utf-8")
print(gdf.head())  # 檢查是否成功讀取
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 讀取 gdf 並過濾該 city 內的所有鄉鎮市區
city_name = "臺北市"  # 你要查看的縣市名稱
city_gdf = gdf[gdf["COUNTYNAME"] == city_name]

# 畫出該縣市的地理邊界
fig, ax = plt.subplots(figsize=(10, 8))
city_gdf.plot(ax=ax, edgecolor="black", alpha=0.5)
ax.set_title(f"{city_name} 地理範圍")
plt.show()
```
